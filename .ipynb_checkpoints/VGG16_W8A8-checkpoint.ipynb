```python
import numpy as np
import torch
import torch.nn as nn
import torchvision
from torchvision import datasets
from torchvision import transforms
from torch.autograd import Variable
from torch.utils.data.sampler import SubsetRandomSampler

# Device configuration
#device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
def data_loader(data_dir,
                batch_size,
                random_seed=42,
                valid_size=0.1,
                shuffle=True,
                test=False):
  
    #normalize = transforms.Normalize(
    #    mean=[0.4914, 0.4822, 0.4465],
    #    std=[0.2023, 0.1994, 0.2010],
    #)
  
    normalize = transforms.Normalize(
        mean=[0.5, 0.5, 0.5],
        std=[0.5, 0.5, 0.5],
    )

    # define transforms
    transform = transforms.Compose([
            transforms.Resize((227,227)),
            transforms.ToTensor(),
            normalize,
    ])

    if test:
        dataset = datasets.CIFAR100(
          root=data_dir, train=False,
          download=True, transform=transform,
        )

        data_loader = torch.utils.data.DataLoader(
            dataset, batch_size=batch_size, shuffle=shuffle
        )

        return data_loader

    # load the dataset
    train_dataset = datasets.CIFAR100(
        root=data_dir, train=True,
        download=True, transform=transform,
    )

    valid_dataset = datasets.CIFAR10(
        root=data_dir, train=True,
        download=True, transform=transform,
    )

    num_train = len(train_dataset)
    indices = list(range(num_train))
    split = int(np.floor(valid_size * num_train))

    if shuffle:
        np.random.seed(random_seed)
        np.random.shuffle(indices)

    train_idx, valid_idx = indices[split:], indices[:split]
    train_sampler = SubsetRandomSampler(train_idx)
    valid_sampler = SubsetRandomSampler(valid_idx)

    train_loader = torch.utils.data.DataLoader(
        train_dataset, batch_size=batch_size, sampler=train_sampler)
 
    valid_loader = torch.utils.data.DataLoader(
        valid_dataset, batch_size=batch_size, sampler=valid_sampler)

    return (train_loader, valid_loader)

batch_size_train = 64
batch_size_test = 1000


# CIFAR100 dataset 
data_loader_train, data_loader_valid = data_loader(data_dir='./data',
                                         batch_size=batch_size_train)

data_loader_test = data_loader(data_dir='./data',
                              batch_size=batch_size_test,
                              test=True)
```

```python
import brevitas.nn as qnn
from brevitas.nn import QuantLinear, QuantReLU, QuantConv2d
import torch.nn as nn
import torch.nn.functional as F
from torch.nn import Module
from brevitas.quant.binary import SignedBinaryActPerTensorConst
from brevitas.quant.binary import SignedBinaryWeightPerTensorConst
from brevitas.inject.enum import QuantType

# Setting seeds for reproducibility
torch.manual_seed(0)

kernel_size=3  
act_bit_width = 8
weight_bit_width = 8

in_channels1=3
out_channels1=64 

in_channels2=64
out_channels2=64

in_channels3=64
out_channels3=128

in_channels4=128
out_channels4=128

in_channels5=128
out_channels5=256

in_channels6=256
out_channels6=256

in_channels7=256
out_channels7=256

in_channels8=256
out_channels8=512

in_channels9=512
out_channels9=512

in_channels10=512
out_channels10=512

in_channels11=512
out_channels11=512

in_channels12=512
out_channels12=512

in_channels13=512
out_channels13=512

input_size = 7*7*512
hidden1 = 4096   
hidden2 = 4096   
num_classes = 100  

class VGG16_W8A8(nn.Module):
    def __init__(self, num_classes=10):
        super(VGG16_W8A8, self).__init__()
        self.input = qnn.QuantIdentity(
                         quant_type='int',
                         scaling_impl_type='const',
                         bit_width=act_bit_width,
                         min_val=-128.0,
                         max_val=127.0, 
                         return_quant_tensor=True
                     )
        
        self.conv1 = qnn.QuantConv2d( 
                         in_channels=in_channels1,
                         out_channels=out_channels1,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn1   = nn.BatchNorm2d(out_channels1)
        self.relu1 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv2 = qnn.QuantConv2d( 
                         in_channels=out_channels1,
                         out_channels=out_channels2,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn2   = nn.BatchNorm2d(out_channels2)
        self.relu2 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        self.pool2 = qnn.QuantMaxPool2d(kernel_size = 2, stride = 2, return_quant_tensor=True)
        
        self.conv3 = qnn.QuantConv2d( 
                         in_channels=out_channels2,
                         out_channels=out_channels3,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn3   = nn.BatchNorm2d(out_channels3)
        self.relu3 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv4 = qnn.QuantConv2d( 
                         in_channels=out_channels3,
                         out_channels=out_channels4,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn4   = nn.BatchNorm2d(out_channels4)
        self.relu4 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        self.pool4 = qnn.QuantMaxPool2d(kernel_size = 2, stride = 2, return_quant_tensor=True)
        
        self.conv5 = qnn.QuantConv2d( 
                         in_channels=out_channels4,
                         out_channels=out_channels5,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn5   = nn.BatchNorm2d(out_channels5)
        self.relu5 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv6 = qnn.QuantConv2d( 
                         in_channels=out_channels5,
                         out_channels=out_channels6,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn6   = nn.BatchNorm2d(out_channels6)
        self.relu6 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv7 = qnn.QuantConv2d( 
                         in_channels=out_channels6,
                         out_channels=out_channels7,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn7   = nn.BatchNorm2d(out_channels7)
        self.relu7 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        self.pool7 = qnn.QuantMaxPool2d(kernel_size = 2, stride = 2, return_quant_tensor=True)
        
        self.conv8 = qnn.QuantConv2d( 
                         in_channels=out_channels7,
                         out_channels=out_channels8,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn8   = nn.BatchNorm2d(out_channels8)
        self.relu8 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv9 = qnn.QuantConv2d( 
                         in_channels=out_channels8,
                         out_channels=out_channels9,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn9   = nn.BatchNorm2d(out_channels9)
        self.relu9 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv10 = qnn.QuantConv2d( 
                         in_channels=out_channels9,
                         out_channels=out_channels10,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn10   = nn.BatchNorm2d(out_channels10)
        self.relu10 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        self.pool10 = qnn.QuantMaxPool2d(kernel_size = 2, stride = 2, return_quant_tensor=True)
        
        self.conv11 = qnn.QuantConv2d( 
                         in_channels=out_channels10,
                         out_channels=out_channels11,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn11   = nn.BatchNorm2d(out_channels11)
        self.relu11 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv12 = qnn.QuantConv2d( 
                         in_channels=out_channels11,
                         out_channels=out_channels12,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn12   = nn.BatchNorm2d(out_channels12)
        self.relu12 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.conv13 = qnn.QuantConv2d( 
                         in_channels=out_channels12,
                         out_channels=out_channels13,
                         kernel_size=kernel_size, 
                         stride=1, 
                         padding=1,
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.bn13   = nn.BatchNorm2d(out_channels13)
        self.relu13 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        self.pool13 = qnn.QuantMaxPool2d(kernel_size = 2, stride = 2, return_quant_tensor=True)
        
        self.drop14 = qnn.QuantDropout(0.5)
        self.fc14   = qnn.QuantLinear(
                         input_size, 
                         hidden1, 
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.relu14 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.drop15 = qnn.QuantDropout(0.5)
        self.fc15   = qnn.QuantLinear(
                         hidden1, 
                         hidden2, 
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        self.relu15 = qnn.QuantReLU(
                         bit_width=act_bit_width, 
                         return_quant_tensor=True
                     )
        
        self.out   = qnn.QuantLinear(
                         hidden2, 
                         num_classes, 
                         weight_bit_width=weight_bit_width,
                         weight_quant_type=QuantType.INT,
                         bias=False
                     )
        
    def forward(self, x):
        out = self.input(x)
        
        out = self.relu1(self.bn1(self.conv1(out)))
        out = self.pool2(self.relu2(self.bn2(self.conv2(out))))
        
        out = self.relu3(self.bn3(self.conv3(out)))
        out = self.pool4(self.relu4(self.bn4(self.conv4(out))))
        
        out = self.relu5(self.bn5(self.conv5(out)))
        out = self.relu6(self.bn6(self.conv6(out)))
        out = self.pool7(self.relu7(self.bn7(self.conv7(out))))
        
        out = self.relu8(self.bn8(self.conv8(out)))
        out = self.relu9(self.bn9(self.conv9(out)))
        out = self.pool10(self.relu10(self.bn10(self.conv10(out))))
        
        out = self.relu11(self.bn11(self.conv11(out)))
        out = self.relu12(self.bn12(self.conv12(out)))
        out = self.pool13(self.relu13(self.bn13(self.conv13(out))))
        
        out = out.reshape(out.size(0), -1)
        
        out = self.relu14(self.fc14(self.drop14(out)))
        out = self.relu15(self.fc15(self.drop15(out)))
        out = self.out(out)
        
        return out
    
model = VGG16_W8A8(num_classes)
```

```python
def train(model, train_loader, optimizer, criterion):
    losses = []
    # ensure model is in training mode
    model.train()    
    
    for i, data in enumerate(train_loader):        
        inputs, target = data
        #inputs, target = inputs.cuda(), target.cuda()
        inputs, target = Variable(inputs), Variable(target)
        
        outputs = model(inputs)
        _,pred = torch.max(outputs.data,1)
        
        optimizer.zero_grad()
        loss = criterion(outputs,target)
 
        loss.backward()
        optimizer.step()
        
        # keep track of loss value
        losses.append(loss.data.numpy()) 
           
    return losses
```

```python
import torch
from sklearn.metrics import accuracy_score

def test(model, test_loader):    
    # ensure model is in eval mode
    model.eval() 
    y_true = []
    y_pred = []
   
    with torch.no_grad():
        for data in test_loader:
            inputs, target = data
            #inputs, target = inputs.cuda(), target.cuda()
            inputs, target = Variable(inputs),Variable(target)
            output = model(inputs)
            #output = torch.sigmoid(output_orig)  
            _,pred = torch.max(output,1)
            # compare against a threshold of 0.5 to generate 0/1
            y_true.extend(target.tolist()) 
            y_pred.extend(pred.reshape(-1).tolist())
        
    return accuracy_score(y_true, y_pred)
```

```python
num_epochs = 10
learn_rate = 0.005 

def display_loss_plot(losses, title="Training loss", xlabel="Iterations", ylabel="Loss"):
    x_axis = [i for i in range(len(losses))]
    plt.plot(x_axis,losses)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
# loss criterion and optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learn_rate, betas=(0.9, 0.999))
```

```python
import numpy as np
from sklearn.metrics import accuracy_score
from tqdm import tqdm, trange

# Setting seeds for reproducibility
torch.manual_seed(0)
np.random.seed(0)

running_loss = []
running_test_acc = []
t = trange(num_epochs, desc="Training loss", leave=True)

for epoch in t:
        loss_epoch = train(model, data_loader_train, optimizer, criterion)
        test_acc = test(model, data_loader_test)
        t.set_description("Training loss = %f test accuracy = %f" % (np.mean(loss_epoch), test_acc))
        t.refresh() # to show immediately the update           
        running_loss.append(loss_epoch)
        running_test_acc.append(test_acc)
```
